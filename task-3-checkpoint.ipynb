```python
# # TASK - 3: Clustering Analysis  Customer Segmentation

# ## Objective:
# Perform customer segmentation using clustering techniques to group customers based on their purchasing behavior (Annual Income and Spending Score). This allows businesses to target each segment effectively.

# ## Required Libraries
import pandas as pd
import numpy as np
import os
os.environ["OMP_NUM_THREADS"] = "1"
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler # For data standardization
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA # For PCA

# ## Step 1: Dataset Selection
# The dataset used is `customer_data.csv`.
# Columns: CustomerID, Genre, Age, Annual_Income_(k$), Spending_Score

# ## Step 2: Tasks to Perform
# ### 1. Load the Dataset
# Load the dataset
df_customer = pd.read_csv("E:\Summer Internship\customer_data.csv")

print("First 5 rows of the customer dataset:")
print(df_customer.head())

# Inspect the dataset
print("\nDataset Shape:")
print(df_customer.shape)

print("\nMissing Values:")
print(df_customer.isnull().sum())

print("\nDuplicate Rows:")
print(df_customer.duplicated().sum()) # Should be 0 for this dataset

print("\nData Types:")
print(df_customer.dtypes)

print("\nSummary Statistics:")
print(df_customer.describe())

# **Observations from Initial Inspection:**
# - The dataset contains 200 rows and 5 columns.
# - There are no missing values.
# - There are no duplicate rows.
# - `CustomerID` is an integer, `Genre` is an object (string), `Age`, `Annual_Income_(k$)`, and `Spending_Score` are integers.
# - The summary statistics give us a good idea of the ranges for Age, Annual Income, and Spending Score.
#
# For clustering, we are primarily interested in `Annual_Income_(k$)` and `Spending_Score`. We might also consider `Age`. `Genre` could be used for post-cluster analysis. `CustomerID` is an identifier and should not be used in clustering.
# ### 2. Data Preprocessing

# **Select features for clustering.**
# For this task, the primary focus is often on 'Annual_Income_(k$)' and 'Spending_Score' as they directly relate to purchasing behavior. We'll start with these two.
# Selecting features for clustering
# Let's rename columns for easier use
df_customer.rename(columns={'Annual_Income_(k$)': 'Annual_Income',
                           'Spending_Score': 'Spending_Score_(1-100)'}, # Assuming it's 1-100 from common versions
                           inplace=True)

# Features for clustering
features = ['Annual_Income', 'Spending_Score_(1-100)']
X = df_customer[features]

print("\nSelected features for clustering (first 5 rows):")
print(X.head())

# **Standardize the data:**
# Use a scaler (e.g., StandardScaler) to ensure all features are on the same scale. This is crucial since clustering algorithms like K-Means are distance-based and sensitive to feature magnitudes.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert scaled data back to a DataFrame for easier inspection (optional)
X_scaled_df = pd.DataFrame(X_scaled, columns=features)

print("\nScaled features (first 5 rows):")
print(X_scaled_df.head())

# ### 3. Clustering

# **Determine the optimal number of clusters:**
# - Use the **Elbow Method**: Plot the Within-Cluster Sum of Squares (WCSS) against the number of clusters. The "elbow point" indicates the optimal number.
# - Alternatively, use the **Silhouette Score** for evaluation.

# Elbow Method to find optimal k
wcss = [] # Within-Cluster Sum of Squares
silhouette_scores = []
possible_k_values = range(1, 11) # Test k from 1 to 10

for k in possible_k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_) # inertia_ is the WCSS for scikit-learn's KMeans

    # Silhouette score requires at least 2 clusters
    if k > 1:
        score = silhouette_score(X_scaled, kmeans.labels_)
        silhouette_scores.append(score)

# Plot WCSS (Elbow Method)
plt.figure(figsize=(10, 5))
plt.plot(possible_k_values, wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS (Inertia)')
plt.xticks(possible_k_values)
plt.grid(True)
plt.show()

# Plot Silhouette Scores
plt.figure(figsize=(10, 5))
# Adjust x-axis for silhouette scores as they start from k=2
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--')
plt.title('Silhouette Score for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(range(2, 11))
plt.grid(True)
plt.show()

# Based on the Elbow Method plot (typically around k=5 for this dataset) and Silhouette Score (highest score)
# k=5 often appears optimal for this specific dataset based on Annual Income and Spending Score.
optimal_k = 5
print(f"\nBased on visual inspection of Elbow method and Silhouette scores, an optimal k is often around {optimal_k}.")

# **Apply K-Means Clustering:**
# - Use the optimal number of clusters identified.
# - Assign a cluster label to each customer.

# Code Cell:
# Apply K-Means with the chosen optimal_k
kmeans_final = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42, n_init=10)
cluster_labels = kmeans_final.fit_predict(X_scaled)

# Add cluster labels to the original DataFrame
df_customer['Cluster'] = cluster_labels

print(f"\nK-Means clustering applied with k={optimal_k}.")
print("First 5 rows with cluster labels:")
print(df_customer.head())

# ### 4. Visualization

# **Create visualizations to represent the clusters:**
# - **2D Scatter Plot:**
#   - Since we used 2 features ('Annual_Income', 'Spending_Score_(1-100)'), we can directly plot them.
#   - If more features were used, PCA or t-SNE would be needed for 2D visualization.
# - **Pair Plots:** Visualize relationships between features within clusters.
# - **Centroid Visuals:** Show the centroid of each cluster.

# 1. 2D Scatter Plot of Clusters
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Annual_Income', y='Spending_Score_(1-100)', hue='Cluster', data=df_customer, palette='viridis', s=100, alpha=0.8, legend='full')

# Plot centroids
centroids = scaler.inverse_transform(kmeans_final.cluster_centers_) # Transform centroids back to original scale
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, color='red', label='Centroids', edgecolor='black')

plt.title(f'Customer Segments (k={optimal_k}) based on Income and Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.grid(True)
plt.show()

# 2. Pair Plots (can include 'Age' and 'Genre' for further exploration)
# We'll create a temporary DataFrame with scaled features and cluster for pairplot
df_for_pairplot = X_scaled_df.copy()
df_for_pairplot['Cluster'] = cluster_labels
df_for_pairplot['Age_Scaled'] = StandardScaler().fit_transform(df_customer[['Age']]) # Scale Age for consistency if used
# Considering the main features used for clustering for pairplot
# If we include 'Age', it gives a sense of how age distributes across income/spending clusters
if 'Age' in df_customer.columns:
    pairplot_features_with_cluster = features + ['Age', 'Cluster']
    plt.figure(figsize=(12,10)) # Ensure a figure is explicitly created for pairplot
    sns.pairplot(df_customer[pairplot_features_with_cluster], hue='Cluster', palette='viridis', diag_kind='kde')
    plt.suptitle('Pair Plot of Features by Cluster', y=1.02)
    plt.show()
else:
    plt.figure(figsize=(10,8))
    sns.pairplot(df_for_pairplot[features + ['Cluster']], hue='Cluster', palette='viridis', diag_kind='kde')
    plt.suptitle('Pair Plot of Features by Cluster', y=1.02)
    plt.show()

# 3. Centroid Visuals (already plotted on the 2D scatter plot)
# We can also list the centroid coordinates
print("\nCluster Centroids (Original Scale):")
centroids_df = pd.DataFrame(centroids, columns=features)
print(centroids_df)



# ## Deliverables

# ### 1. Clustered Dataset:
# - A new column 'Cluster' has been added to `df_customer` with the assigned cluster labels for each customer.

# Code Cell:
print("\n--- Clustered Dataset ---")
print("DataFrame with Cluster Labels (first 10 rows):")
print(df_customer.head(10))

# ### 2. Visualizations:
# - **A 2D scatter plot representing customer clusters:** Generated above, showing segments based on Annual Income and Spending Score.
# - **Elbow Method plot showing the optimal number of clusters:** Generated above.
# - **Any additional visuals (e.g., pair plots):** Pair plots were generated above to show relationships between features within clusters.

# ### 3. Recommendations:
# Insights and actions based on clusters. (These are general interpretations based on typical results for k=5 with this dataset. Exact centroid values and cluster characteristics might vary slightly).

# Assuming 5 clusters are formed, they typically represent:
# *   **Cluster 0 (e.g., Low Income, Low Spending): "Careful/Standard"**
#     *   _Characteristics:_ Modest annual income and a low spending score.
#     *   _Recommendations:_ Target with budget-friendly offers, basic necessities, or small-value promotions. Loyalty programs might not be highly effective unless very low-threshold.
# *   **Cluster 1 (e.g., High Income, Low Spending): "Savers/Target for Upsell"**
#     *   _Characteristics:_ High annual income but a low spending score. They earn well but are cautious spenders.
#     *   _Recommendations:_ Target with premium products that emphasize value, durability, or long-term benefits. Exclusive offers or personalized services might entice them. Understand why they are not spending  perhaps they need convincing of product quality or unique value.
# *   **Cluster 2 (e.g., Mid Income, Mid Spending): "Average/Balanced"**
#     *   _Characteristics:_ Average annual income and an average spending score. This is often the largest group.
#     *   _Recommendations:_ General marketing campaigns, promotions on popular items, loyalty programs with achievable rewards. They are susceptible to good deals and variety.
# *   **Cluster 3 (e.g., Low Income, High Spending): "Enthusiasts/Impulse Buyers (Potentially Risky)"**
#     *   _Characteristics:_ Low annual income but a high spending score. They might be younger, students, or impulse buyers.
#     *   _Recommendations:_ Target with trendy items, flash sales, and products that offer instant gratification. Be mindful of credit risk if offering payment plans. Loyalty programs focused on frequency could work.
# *   **Cluster 4 (e.g., High Income, High Spending): "Prime/VIP Customers"**
#     *   _Characteristics:_ High annual income and a high spending score. These are the ideal customers.
#     *   _Recommendations:_ Target with premium products, exclusive access, personalized services, and high-value loyalty programs. Make them feel valued. Early access to new products.

# **Further Tailored Marketing Strategies:**
# - **Age & Genre:** Once clusters are defined by income/spending, analyze the distribution of `Age` and `Genre` within each cluster.
#   - _Example:_ If "Prime Customers" are predominantly in a certain age group, tailor marketing visuals and channels accordingly.
# - **Promotions:** Design promotions specific to the needs and price sensitivity of each segment.
# - **Loyalty Programs:**
#   - For "Prime Customers": Tiered programs with exclusive benefits.
#   - For "Average Customers": Points-based systems for discounts.
#   - For "Enthusiasts": Rewards for frequent, smaller purchases.
```
